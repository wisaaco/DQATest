```python
import gradio as gr
import time
def greet(Description, File, progress=gr.Progress()):
    progress(0, desc="Starting")
    # return "Hello, " + Description + "!" * int(intensity) +" "+ texto 
    progress(0.05)
    response = []
    for i in range(3):
        time.sleep(0.3)
        response.append("Hello, %i\n"%i)
        yield " ".join(response)

demo = gr.Interface(
    fn=greet,
    # inputs=["textbox", "slider","text","image"],
    inputs=["textbox", "file"],
    outputs=["text"],
).queue(default_concurrency_limit=5)

demo.launch()
# demo.launch(share="True")
```
